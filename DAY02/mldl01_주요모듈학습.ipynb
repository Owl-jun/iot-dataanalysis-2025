````python
# 모듈 사용
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import rcParams, font_manager, rc
font_path = 'C:/Windows/Fonts/malgun.ttf' # 맑은고딕체로 변경
font = font_manager.FontProperties(fname=font_path).get_name() # 실제 설치된 폰트 이름조회
rc('font', family=font) # 한글깨짐현상 해결!!
rcParams['axes.unicode_minus'] = False # 한글 사용시 마이너스 표시 깨짐 해결!
````

````python
# 기본데이터

arr = np.random.random(30)
arr
````

````python
# 단일차트
plt.plot(arr)
plt.show()
````

````python
# 히스토리 맵
plt.hist(np.random.randn(100))
plt.show()
````

````python
# 산점도
plt.scatter(np.arange(30),np.random.randn(30))
plt.show()
````

````python
# 다중차트 - 여러차트를 한꺼번에 출력
fig = plt.Figure(figsize=(15,8)) # 객체 생성

ax1 = fig.add_subplot(2, 2, 1)  # 2x2 차트 중 1번째
ax2 = fig.add_subplot(2, 2, 2)  # 2x2 차트 중 1번째
ax4 = fig.add_subplot(2, 2, 4)  # 2x2 차트 중 1번째

ax1.hist(np.random.randn(100))
ax2.scatter(np.arange(30),np.random.randn(30))
ax4.plot(arr)
fig
````

````python
# 다중차트 - 여러차트를 한꺼번에 출력
# 차트 꾸미기
fig, axes = plt.subplots(2,2, figsize=(15,8))

# color 색상변경, marker 포인트 모양 (o, ^, *, 1)
axes[0][0].hist(np.random.randn(100), color = 'green')
axes[0][1].scatter(np.arange(30),np.random.randn(30), color = 'blue', marker = '^')
axes[1][0].plot(arr, color = 'yellow', linestyle = '-', marker = 'o')
axes[1][1].plot(arr, linestyle = '--')
plt.show()
````

````python
# 모듈 사용
import folium
import json
from folium import Marker
````

````python
# 서울 중심 위치
# 위 경도는 소수점 4자리 이상으로 사용해야 오차가 적다.
loc = (37.5776087830657, 126.976896737645)

json_data = open("./seoul_municipalities_geo_simple.json",encoding='utf-8').read()
data = json.loads(json_data)

m = folium.Map(location=loc, zoom_start=9)
folium.GeoJson(data, name='동').add_to(m)
Marker(location=loc,popup='경복궁', icon=folium.Icon(color='lightblue',icon='stop')).add_to(m)
m

# folium.Icon 종류 - 
````
